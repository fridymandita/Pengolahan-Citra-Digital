# Load Library

```python
import cv2
import matplotlib.pyplot as plt
```

#Convert Image ke RGB

```python
img = cv2.imread('Scenery.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

#Baca Image ke Grayscale

```python
gray = cv2.imread('Scenery.png', 0)
```

#Generate Negative Image

```python
colored_negative = abs(255-img)
gray_negative = abs(255-gray)
```

#Plot Result Negative Image

```python
imgs = [img, gray, colored_negative, gray_negative]
title = ['coloured', 'gray', 'coloured-negative', 'gray-negative']


plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title(title[1])
plt.imshow(imgs[1], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 3)
plt.title(title[2])
plt.imshow(imgs[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 4)
plt.title(title[3])
plt.imshow(imgs[3], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```
